```python
import pandas as pd
import joblib
import numpy as np
```

```python
train_df = pd.read_csv('../../datasets/train_set.csv', sep='\t')
test_df = pd.read_csv('../../datasets/test_a.csv', sep='\t')
```

```python
load_vocal = joblib.load('../../intermediate_save_data/vocal.pkl')
print(load_vocal.get_stoi())
```

```python
print(load_vocal.get_stoi().get('349', 0))
print(load_vocal.get_stoi().get('3113', 0))
print(load_vocal.get_stoi().get('4806', 0))
```

```python
%%time

def split_truncate_pad(string,
                       num_steps,  # 
                       stoi,  # Dictionary mapping tokens to indices.
                       padding_index):  # '<pad>'
    """"""
    # Vocab
    line = [stoi.get(word, 0) for word in string.split()]
    if len(line) > num_steps:
        # 
        return line[:num_steps]  # 
    return line + [padding_index] * (num_steps - len(line))  # 


X_train_data = train_df['text'].apply(split_truncate_pad, num_steps=3000,
                                       stoi=load_vocal.get_stoi(), padding_index=1)
X_test_data = test_df['text'].apply(split_truncate_pad, num_steps=3000, stoi=load_vocal.get_stoi(),
                                    padding_index=1)
```

```python
# 
joblib.dump(np.array(X_train_data.values.tolist()), '../../intermediate_save_data/X_train.pkl')
joblib.dump(train_df['label'].values, '../../intermediate_save_data/y_train.pkl')
joblib.dump(np.array(X_test_data.values.tolist()), '../../intermediate_save_data/X_test.pkl')
```
