```python
import pandas as pd
from gensim.models.word2vec import Word2Vec
```

```python
train_dataset = pd.read_csv('../../datasets/train_set.csv', sep='\t')['text']
test_dataset = pd.read_csv('../../datasets/test_a.csv', sep='\t')['text']
```

```python
# 训练数据格式为列表
all_dataset = pd.concat([train_dataset, test_dataset]).str.split().values.tolist()  # 全部的可训练数据(合并训练与测试)
```

```python
model = Word2Vec(all_dataset, workers=8, vector_size=300, epochs=15,
                 window=5)  # 默认window=5
model
```

```python
model.wv['669']  # 单词'669'代表的词向量
```

```python
# 保存训练好的词向量
model.wv.save_word2vec_format("cnew_300.txt")
```
